```python
import pandas as pd

# Open and read a .txt file
file_path = 'EXAMPL_DATA.txt'  # Change this to your file path

# Read the file into a DataFrame (assuming space-separated values)
df = pd.read_csv(file_path, delim_whitespace=True, header=None)

# Display the content and number of columns
print("File Content:")
print(df)
print(f"Number of columns: {df.shape[1]}")


#Python script that moves column 5 after column 3 and removes column 6:


# Open and read a .txt file
file_path = 'EXAMPL_DATA.txt'  # Change this to your file path
new_file_path = 'modified_file.txt'  # New file name

# Read the file into a DataFrame (assuming space-separated values)
df = pd.read_csv(file_path, delim_whitespace=True, header=None)

# Move column 5 (index 4) after column 3 (index 2)
cols = list(df.columns)
cols.insert(3, cols.pop(4))  # Move column at index 4 to index 3
df = df[cols]

# Remove column 6 (now at index 5 after the move)
if len(df.columns) > 5:
    df.drop(df.columns[5], axis=1, inplace=True)

# Save the modified DataFrame to a new text file
df.to_csv(new_file_path, sep='\t', index=False, header=False)

# Read and display the modified file
with open(new_file_path, 'r', encoding='utf-8') as file:
    modified_data = file.read()

print("Modified File Content:")
print(modified_data)


#script that processes modified_file.txt, splits it based on unique values in column 1, and saves each group into a separate file named after the unique values in column 1:
#Reads modified_file.txt.
#Extracts unique values from column 1.
#Creates separate .txt files for each unique value in column 1.
#Removes column 1 from the saved files.
#Try using a tab ('\t') instead of a space (' ') to make the output more readable. 
# A tab (\t) provides more consistent spacing than a single space, making the columns easier to read.

# Read the modified file
file_path = 'modified_file.txt'  # Input file

# Read the file into a DataFrame (assuming space-separated values)
df = pd.read_csv(file_path, delim_whitespace=True, header=None)

# Get unique values from column 1 (index 0)
unique_values = df[0].unique()

# Process each unique value
for value in unique_values:
    subset = df[df[0] == value].iloc[:, 1:]  # Remove column 1
    output_file = f"{value}.txt"  # Name file based on column 1 value
    subset.to_csv(output_file, sep='\t', index=False, header=False)  # Use tab separator

print("Files saved successfully.")
```
